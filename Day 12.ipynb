```python
dataset=pd.read_csv("general_data.csv")
dataset1=dataset[dataset["Attrition"] =="Yes"]
dataset1.head()
```

```python
# wilcoxon test
from scipy.stats import wilcoxon
stats,p=wilcoxon(dataset1.TrainingTimesLastYear,dataset1.YearsAtCompany)
print(stats,p)
if p<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
# friedman test
from scipy.stats import friedmanchisquare
stats,p=friedmanchisquare(dataset1.TrainingTimesLastYear,dataset1.YearsAtCompany,dataset1.YearsSinceLastPromotion)
print(stats,p)
if p<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
dataset=pd.read_csv("general_data.csv")
dataset2=dataset[dataset["Attrition"] =="No"]
dataset2.head()
```

```python
# mannwhitney test
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(dataset1.DistanceFromHome,dataset2.DistanceFromHome)
print(stats,p)
#Null Hypothesis:there is no significance difference between attrition and distance from home.
if p<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
# kruskal willis test
from scipy.stats import kruskal
stats,p=kruskal(dataset1.DistanceFromHome,dataset2.DistanceFromHome,dataset1.Education)
print(stats,p)
#Null Hypothesis:there is no significance difference between attrition, distance from home and education.
if p<0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

```python
# chisquare
dataset3=dataset.dropna()
dataset3.head()
```

```python
from scipy.stats import chi2_contingency
chitable=pd.crosstab(dataset3.Attrition,dataset3.Gender)
chitable
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
print(stats,p)
#There is no significance difference between attrition and gender
if p<0.05:
    print("Reject Null hypothesis")
else:  
    print("Accept Null hypothesis")
```

```python
# one sample test
dataset1["Education"]
```

```python
from scipy.stats import ttest_1samp
stats,p=ttest_1samp(dataset1.Education,5)
print(stats,p)
#there is no significance difference between mean of Education and population mean=5
if p<0.05:
    print("Reject Null hypothesis")
else:
    print("Accept Null hypothesis")
```

```python
# two sample paired testprint
(dataset1["DistanceFromHome"],dataset1["Education"])
```

```python
from scipy.stats import ttest_rel
stats,p=ttest_rel(dataset1.Education,dataset1.DistanceFromHome)
print(stats,p)
#there is no significance difference between mean of Education and mean of distance from home.
if p<0.05:
    print("Reject Null hypothesis")
else:
    print("Accept Null hypothesis")
```

```python
# Two sample independent test
dataset4=dataset.loc[dataset["Attrition"]=="Yes"]
dataset4.Attrition=dataset4.Attrition.map({"Yes":1})
dataset4.head()
```

```python
dataset5=dataset.loc[dataset["Attrition"]=="No"]
dataset5.Attrition=dataset5.Attrition.map({"No":0})
dataset5.head()
```

```python
from scipy.stats import ttest_ind
stats,p=ttest_ind(dataset4.Education,dataset5.Education)
print(stats,p)
#there is no significance difference between Attrition and mean of Education with respect to attrition yes or no.
if p<0.05:
    print("Reject Null hypothesis")
else:
    print("Accept Null hypothesis")
```
